```python
import os 
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx  
plt.style.use('ggplot') 
%matplotlib inline 
```

```python
fig = plt.figure(figsize=(10,6))  #Set the size of te figure 
ax = fig.gca() #Get the current axes of the figure 
x = np.linspace(-1, 1, 200) # Return evenly spaces numbers over a given range 
hz =10
cost = np.sin(hz*x)*np.exp(-x)
cost2 = np.exp(-x)
cost3 = np.sin(hz*x)

ax.plot(x,cost) #Plot the function cost1, with axis data=x
ax.plot(x,cost2)
ax.plot(x,cost3)
ax.set_ylabel('Cost')
ax.set_xlabel('Some Parameter')
```

```python
fig = plt.figure(figsize=(10,6))  #NEEDS to be defined for every cell box!! 
gradient = np.diff(cost) #Calculate the first difference (or the nth discerte difference along some given axis)
bx = fig.gca()
bx.plot(x[0:199], gradient*5 ) -> scaled value of difference...equivalent to a delta
bx.plot(x, cost )
```
